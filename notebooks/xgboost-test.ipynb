```python
import xgboost as xgb
import pandas as pd
```

```python
data = pd.read_parquet("../data/2022-2023/data.parquet")
data.drop(columns="game_id", inplace=True)
data
```

```python
X = data.drop(columns="is_goal")
y = data["is_goal"]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, average_precision_score
```

```python
sc = StandardScaler()
```

```python
X_scaled = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=0, test_size=0.2)
```

```python
model = xgb.XGBClassifier(n_jobs=-1, n_estimators=5000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Test", accuracy_score(y_pred, y_test), "Train", accuracy_score(model.predict(X_train), y_train), "ROC AUC Score", roc_auc_score(y_test, y_pred), "PR AUC Score", average_precision_score(y_test, y_pred))
```
